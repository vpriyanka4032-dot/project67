```python
#CAPPING
df["Fare_capped"]=df["Fare"]
df.loc[df["Fare_capped"]>8.8,"Fare_capped"]=8.8
df.loc[df["Fare_capped"]<5.11,"Fare_capped"]=5.11
df.head()
```

```python
#IQR = Q3-Q1
q1 = df["Fare_capped"].quantile(0.25)
print(q1)
q3 = df["Fare_capped"].quantile(0.75)
print(q3)
```

```python
iqr = q3-q1
print(iqr)
```

```python
upper_fence = q3+1.5 * iqr
lower_fence = q1-1.5 * iqr
```

```python
print("upper fence:",upper_fence)
print("lower fence:",lower_fence)
```

```python
# trimmed data
plt.figure(figsize=(10,8))
plt.subplot(2,2,1)
sns.distplot(df["Fare_capped"])

plt.subplot(2,2,2)
sns.boxplot(df["Fare_capped"])

plt.subplot(2,2,3)
sns.distplot(df["Fare_capped"])

plt.subplot(2,2,4)
sns.boxplot(df["Fare_capped"])
plt.tight_layout()
plt.show()
```

```python
# trimming
new_df_1 =df[(df["Fare_capped"]<upper_fence) & (df["Fare_capped"]>lower_fence)]
new_df_1
```

```python
# compairing original data vs trimmed data
plt.figure(figsize=(10,8))
plt.subplot(2,2,1)
sns.distplot(df["Fare_capped"])

plt.subplot(2,2,2)
sns.boxplot(df["Fare_capped"])

plt.subplot(2,2,3)
sns.distplot(new_df_1["Fare_capped"])

plt.subplot(2,2,4)
sns.boxplot(new_df_1["Fare_capped"])
plt.tight_layout()
plt.show()
```

```python
features selections:
Feature Selection means choosing only the most relevant and useful columns (features) from your dataset that contribute the most to your model’s prediction accuracy.
types of features selections
1. filter method:
       These methods select features based on statistical tests between independent (X) and dependent (y) variables.
i) chisquare test:
         The Chi-Square (χ²) test is a statistical test used to determine if there is a relationship between two categorical variables
ii) anova test:
         ANOVA checks whether there’s a significant difference between the means of different groups.    
iii) correlation coefficient:
         The correlation coefficient (r) measures how strongly two numerical variables are related.
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
import warnings
warnings.filterwarnings("ignore")
```

```python
# load titanic data
df = sns.load_dataset("titanic")
print(df.shape)
df.head()
```

```python
df.isnull().sum()
```

```python
df.drop(["deck"],axis=1,inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df.info()
```

```python
le = LabelEncoder()
df["sex"]=le.fit_transform(df["sex"])
df["class"]=le.fit_transform(df["class"])
df["alone"]=le.fit_transform(df["alive"])
df["embarked"]=le.fit_transform(df["embarked"])
df["who"]=le.fit_transform(df["who"])
df["adult_male"]=le.fit_transform(df["adult_male"])
df["embark_town"]=le.fit_transform(df["embark_town"])
df["alive"]=le.fit_transform(df["alive"])
df
```

```python
x=df.drop("survived",axis=1) # independent variable
y=df["survived"] # target values
```

```python
# chisquare test
from sklearn.feature_selection import SelectKBest,chi2,f_classif   # k - no of features .important

chi2_selector = SelectKBest(score_func=chi2,k="all")
chi2_selector.fit(x,y)
chi2_selector.scores_
```

```python
import numpy as np
chi2_scores_1 = pd.DataFrame({
    "columns_1":x.columns,
    "chi2_scores":np.round(chi2_selector.scores_,2)
})
chi2_scores_1
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x="chi2_scores",y="columns_1",data=chi2_scores_1,palette="viridis")   # dark purple → blue → green → yellow.
plt.show()
```

```python
# Anova
from sklearn.feature_selection import SelectKBest,chi2,f_classif
anova_selector=SelectKBest(score_func=f_classif,k="all")
anova_selector.fit(x,y)
anova_selector.scores_
```

```python
import numpy as np
anova_scores_1 = pd.DataFrame({
    "columns_1":x.columns,
    "anova_scores_1":np.round(anova_selector.scores_,2)
})
anova_scores_1
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x="anova_scores_1",y="columns_1",data=anova_scores_1,palette="magma")  #yellow → red → dark
plt.show()
```
